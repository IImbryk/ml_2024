```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(random_state=42)
```

```python
df2 = df.drop(['state', 'phone number'], axis=1)
```

```python
df2['international plan'] = df2['international plan'].map({'yes': 1, 'no': 0})
```

```python
df2['voice mail plan'] = df2['voice mail plan'].map({'yes': 1, 'no': 0})
```

```python
tsne.fit_transform(df2.drop(['churn'], axis=1))
```

```python
tsne_repr = tsne.embedding_
```

```python
plt.scatter(tsne_repr[:, 0], tsne_repr[:, 1])
```

```python
plt.scatter(tsne_repr[~df['churn'].values, 0],
            tsne_repr[~df['churn'].values, 1], color='blue')

plt.scatter(tsne_repr[df['churn'].values, 0],
            tsne_repr[df['churn'].values, 1], color='orange')
```

```python
_, axes = plt.subplots(1, 2, sharey=True, figsize=(16,6))

axes[0].scatter(tsne_repr[:, 0], tsne_repr[:, 1], 
            c=df['international plan'].map({'yes': 'blue', 'no': 'orange'}));

# axes[1].scatter(tsne_repr[:, 0], tsne_repr[:, 1], 
#             c=df['voice mail plan'].map({'yes': 'blue', 'no': 'orange'}));

axes[1].scatter(tsne_repr[:, 0], tsne_repr[:, 1], 
            c=df['voice mail plan'].map({'yes': 'blue', 'no': 'orange'}));

axes[0].set_title('international plan');
axes[1].set_title('voice mail plan');
```
